# **Series**

```
L1 = [0, 1, 2, 5, 10]
s1=pd.Series(L1)
print(s1)
print(type(s1))
```

```
import numpy as np
nm=np.random.randn(5)

s2=pd.Series(nm)
s2
```

```
dct={'a':1, 'b':2, 'c':3, 'd':4,'e':5}
s3=pd.Series(dct)
print(s3)
```

```
print(s1)  # Changing index
s1.index=['a','b','c','d','e']
print(s1)
```

```
# slicing
print(s1[2:])
print(s1[-2])
print(s1['a':'c'])
```

```
s4=s1.append(s2)
print(s4)
```

```
print(s4.drop('d'))
```

```
L1 = [0, 1, 2, 5, 10]
L2 = [0, 3, 6, 7, 8, 9, 56]
s1=pd.Series(L1)
s2=pd.Series(L2)

s2.add(s1)  # also sub, mul, div
```

```
print(s2.mean())
print(s2.median())

# also max, min 
```

```
print(s2.describe())
```

# **Dataframe**

```
dates=pd.date_range('today', periods=6)
arr=np.random.randn(6,4)
cols=['a','b','c','d']
df1 = pd.DataFrame(arr, index=dates, columns=cols)
df1
```

```
df1.head(2) #returns first 5 by default, tail last 5 
```
