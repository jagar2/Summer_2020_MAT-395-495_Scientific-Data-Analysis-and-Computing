# Joining, Splitting, Resizing

```
a=np.array([[1,2],[3,4]])
print(a)
b=np.array([[5,6],[7,8]])
print(b)
```

```
print(np.concatenate((a,b)))
```

```
print(np.concatenate((a,b), axis = 1))  # default axis 0 (down)
```

```
a=np.arange(9)
print(a)
print(np.split(a,3))
```

```

print(np.split(a,[3,6]))

print(np.split(a,[4,7]))
```

```
# Resize
a = np.array([[1,2,3],[4,5,6]])
print(a)
print(a.shape)
b=np.resize(a, (3,2))
print(b)
print(b.shape)
```

```
c=np.resize(a,(3,3))
print(c)
```

```
a=np.linspace(1,3,10)
print(a)
a=np.logspace(10, 15,6)
print(a)
```

```
a=np.array([(1,2,3), (10, 10, 10)])
print(a)

print("sum down axis 0 is ",a.sum(axis=0))
```

```
print(np.sqrt(a))
print(np.std(a))
```

```
print(np.log(a))
print(np.log10(a))
```

# Examples

```
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,6*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show
```

```
a=np.random.rand(10,10)
a[np.random.randint(10,size=5), np.random.randint(10,size=5)] = np.nan
print("total number of missing values \n", np.isnan(a).sum())
print("nans are at locations:", np.argwhere(np.isnan(a)))
```

```
i=np.where(np.isnan(a))
a[i]=0
print(a)
```
