# **Series**

```
import pandas as pd
```

```
L1= [0, 1, 2, 4, 6]
s1 = pd.Series(L1)
print(s1)
print(type(s1))
```

```
import numpy as np
nm=np.random.randn(5)

s2=pd.Series(nm)
s2
```

```
dct={'a':1, 'b':2, 'c':3, 'd':4, 'e':8}
s3=pd.Series(dct)
s3
```

```
L1= [0, 1, 2, 4, 6]
s1 = pd.Series(L1)
print(s1)
s1.index=['a','b','c','d','e']
print(s1)
```

```
#slicing
print(s1[2:])
print(s1[-2])
print(s1['a':'c'])
```

```
L1 = [0, 1, 3, 5, 11]
L2 = [0, 3, 5, 7, 7, 7, 45]
s1 = pd.Series(L1)
s2 = pd.Series(L2)

s2.add(s1)
# also  sub, mul, div
```
