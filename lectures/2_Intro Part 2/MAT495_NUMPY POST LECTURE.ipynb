# **Array Manipulation**

```
a=np.arange(9)
print(a)
```

```
b=a.reshape(3,3)
print(b)
```

```
print(b.flatten())
```

```
a = np.arange(12).reshape(2,6)
print(a)
```

# **Slice**

```
a=np.arange(20)
print(a)
print(a[4:])
```

```
print(a[:5])
```

```
slc=slice(2, 15, 1)
b=a[slc]
print(b)
print(type(slc))
```

# **Joinging, Splitting, Resizing**

```
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
print(a)
print(b)
```

```
print(np.concatenate((a,b), axis=1))
```

```
print(np.concatenate((b,a), axis=0))
```

```
c=np.concatenate((b,a), axis=0)
print(a.shape)
print(c.shape)
```

```
a=np.array([[1,2,3],[4,5,6]])
print(a)
print(a.shape)

b=np.resize(a, (3,2))
print(b)
```

```
c=np.resize(a,(3,3))
print(c)
```

```
a=np.linspace(1,3,10, dtype=float)
print(a)
a=np.logspace(10,15,11)
print(a)
```

```
a=np.array([[1,2,3],[10,10,10]])
print(a)
print("sum down axis 0 is ", a.sum(axis=0))
```

```
print(np.sqrt(a))
print(np.std(a))
print(np.log(a))
```
