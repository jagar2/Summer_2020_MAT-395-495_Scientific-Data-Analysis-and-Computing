```
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

# Problem 1

1) Plot the binomial function using both the built-in Python function and directly using mathematical expression for n = 10, p = 1/6. Remember the binomial is a *discrete* function.

## Solution

```
# define constants
n = 10
p = 1/6
x_min = 0
x_max = 10

x = np.linspace(x_min,x_max,x_max+1)
# Note that using (np or sp).math.factorial requires a scalar input. The special can take arrays
y_manual = sp.special.factorial(n)/(sp.special.factorial(x)*sp.special.factorial(n-x))*p**x*(1-p)**(n-x)
y_function = stats.binom.pmf(x, n, p)

# build the plot
fig = plt.figure(figsize =(6,10))
ax1 = fig.add_subplot(311)
ax1.plot(x, y_manual,'o')
ax1.set_xlabel('x value')
ax1.set_ylabel('Binomial Manual')
ax1.set_title('Binomial Curve')

ax2 = fig.add_subplot(312)
ax2.plot(x, y_function, 'o')
ax2.set_xlabel('x value')
ax2.set_ylabel('Binomial Function')

ax3 = fig.add_subplot(313)
ax3.plot(x, y_manual - y_function, 'o')
ax3.set_xlabel('x value')
ax3.set_ylabel('Binomial Difference')

plt.tight_layout()
plt.show()
```

# Problem 2

2) Plot the Poission function using both the built-in Python function and directly using mathematical expression for μ = 2. Remember the binomial is a *discrete* function.

## Solution

```
# define constants
mu = 2
x_min = 0
x_max = 10

x = np.linspace(x_min,x_max,x_max+1)
# Note that using (np or sp).math.factorial requires a scalar input. The special can take arrays
y_manual = mu**x/sp.special.factorial(x)*np.exp(-mu)
y_function = stats.poisson.pmf(x,mu)

# build the plot
fig = plt.figure(figsize =(6,10))
ax1 = fig.add_subplot(311)
ax1.plot(x, y_manual,'o')
ax1.set_xlabel('x value')
ax1.set_ylabel('Poisson Manual')
ax1.set_title('Poisson Curve')

ax2 = fig.add_subplot(312)
ax2.plot(x, y_function, 'o')
ax2.set_xlabel('x value')
ax2.set_ylabel('Poisson Function')

ax3 = fig.add_subplot(313)
ax3.plot(x, y_manual - y_function, 'o')
ax3.set_xlabel('x value')
ax3.set_ylabel('Poisson Difference')

plt.tight_layout()
plt.show()
```

# Problem 3


3) Plot the Gaussian function using both the built-in Python function and directly using mathematical expression for μ = 2, σ = 0.1.

## Solution

```
# define constants
mu = 2 
sigma = 0.1
x_min = mu-5*sigma
x_max = mu+5*sigma

x = np.linspace(x_min,x_max,10**2)
y_manual = 1/(sigma * np.sqrt(2* np.pi))*np.exp(-1/2*((x-mu)/sigma)**2)
y_function = stats.norm.pdf(x, mu, sigma)

# build the plot
fig = plt.figure(figsize =(6,10))
ax1 = fig.add_subplot(311)
ax1.plot(x, y_manual)
ax1.set_xlabel('x value')
ax1.set_ylabel('Gaussian Manual')
ax1.set_title('Gaussian Curve')

ax2 = fig.add_subplot(312)
ax2.plot(x, y_function)
ax2.set_xlabel('x value')
ax2.set_ylabel('Gaussian Function')

ax3 = fig.add_subplot(313)
ax3.plot(x, y_manual - y_function)
ax3.set_xlabel('x value')
ax3.set_ylabel('Guassian Difference')

plt.tight_layout()
plt.show()
```

# Problem 4

4) Plot the Cauchy (Lorentzian) function using both the built-in Python function and directly using mathematical expression for μ = 2, FWHM = 10. 

## Solution

```
# define constants
mu = 2 
FWHM = 0.1
x_min = mu-5*FWHM
x_max = mu+5*FWHM

x = np.linspace(x_min,x_max,10**2)
y_manual = 1/np.pi * FWHM/2 / ((x-mu)**2 + (FWHM/2)**2)
y_function = stats.cauchy.pdf(x, mu, FWHM/2)

# build the plot
fig = plt.figure(figsize =(6,10))
ax1 = fig.add_subplot(311)
ax1.plot(x, y_manual)
ax1.set_xlabel('x value')
ax1.set_ylabel('Gaussian Manual')
ax1.set_title('Gaussian Curve')

ax2 = fig.add_subplot(312)
ax2.plot(x, y_function)
ax2.set_xlabel('x value')
ax2.set_ylabel('Gaussian Function')

ax3 = fig.add_subplot(313)
ax3.plot(x, y_manual - y_function)
ax3.set_xlabel('x value')
ax3.set_ylabel('Guassian Difference')

plt.tight_layout()
plt.show()
```
