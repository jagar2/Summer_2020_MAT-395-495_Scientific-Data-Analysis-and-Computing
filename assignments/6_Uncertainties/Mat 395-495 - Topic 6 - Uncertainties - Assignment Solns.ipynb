3)	To check the proportionality of P to T for the ideal gas given a constant volume, a student measures the pressure of the gas at five different temperatures (at the same volumes) and get the results shown below. Plot these results in a graph of P vs. T using Python. Make sure to include error bars for each point and label axis appropriately. Decide whether they confirm the expected proportionality of P and T.


Temperature (K) |  Pressure (atm)
:-:|:-:
100 ± 2	| 0.36 ± 0.04
150 ± 2	| 0.46 ± 0.04
200 ± 2	| 0.71 ± 0.04
250 ± 2	| 0.83 ± 0.04
300 ± 2	| 1.04 ± 0.04


## Solution

```
import numpy as np
import matplotlib.pyplot as plt

# Data set with associated errors
T = np.linspace(100,300,5)
T_error = np.linspace(2,2,5)
P = [0.36, 0.46, 0.71, 0.83, 1.04]
P_error = np.linspace(0.04,0.04,5)

# Ideal gas law states that P/T is a constant
# Calculate and plot to see if fall on a constant value
# Note: analysis of error will be discussed in a future class
Ideal_Gas = P/T
Ideal_Gas_error = np.sqrt((P_error/P)**2 + (T_error/T)**2)*Ideal_Gas

# Plot data
# Experimental data with error bars
fig1, ax1 = plt.subplots() 
ax1.plot(T,P,'o')
ax1.set_xlabel("Temperature (K)")
ax1.set_ylabel("Pressure (atm)")
ax1.errorbar(T,P,yerr=P_error,xerr=T_error, capsize=4, linestyle='')

# Plot calculated, anticipated relationshiop if following ideal gas law
fig2, ax2 = plt.subplots() 
ax2.plot(T,Ideal_Gas,'o')
ax2.set_xlabel("Temperature (K)")
ax2.set_ylabel("P/T (atm/K)")
ax2.errorbar(T,Ideal_Gas,yerr=Ideal_Gas_error, capsize=4, linestyle='')
ax2.plot(T,np.linspace(0.00333,0.00333,len(T)))
```

It would appear that the data indeed follows the ideal gas law as the deviations of the proportionality constant appear to be  random. 4)	To measure the effectiveness of a scintillator material you synthesized, you expose it to a radioactive source. (A scintillator material absorbs high energy particles, e.g. as emitted during nuclear fission, and emits a light pulse as a response.) You task two students to count the number of light pulses the scintillator emits. Student A watches your material for 3 minutes and counts 28 pulses. Student B watches it for 30 minutes and counts 310 pulses. 

1. What should Student A report for the average number emitted pulses in 3 minutes with the appropriate uncertainty?
1. What should Student B report for the average number emitted pulses in 30 minutes with the appropriate uncertainty?
1. What are the fractional uncertainties in the two measurements? Briefly comment on what you learn from this.



## Solution

```
# Data follows a Poisson Distribution, hence the error is sqrt(N)

# Student A:
A_Pulse = 28 #count
A_Time = 3 #min
A_Pulse_err_abs = np.sqrt(A_Pulse)

print("Student A should report:", np.around(A_Pulse,0), "±", np.around(A_Pulse_err_abs,0))

# Student B:
B_Pulse = 310 #count
B_Time = 30 #min
B_Pulse_err_abs = np.sqrt(B_Pulse)

print("Student B should report:", np.around(B_Pulse,0), "±", np.around(B_Pulse_err_abs,0))

# Relative errors for Student A, B:
A_Pulse_err_rel = A_Pulse_err_abs/A_Pulse*100
B_Pulse_err_rel = B_Pulse_err_abs/B_Pulse*100

print("Error for Student A: " , np.around(A_Pulse_err_rel,0), "% and Student B:", np.around(B_Pulse_err_rel,0), "%")
```

<font color='red'>Student A has a notably higher error which is directly related to the shorter period of data collection. </font>
