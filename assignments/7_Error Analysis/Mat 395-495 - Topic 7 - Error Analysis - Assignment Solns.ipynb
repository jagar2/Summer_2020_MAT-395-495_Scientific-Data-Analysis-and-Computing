# Problem 1

1) You make the following measurements as part of characterizing your sample in the lab:

a =	5 ± 1 cm, 

b = 1.7 ± 0.2 cm, 

c =	12 ± 1 cm, 

t =	3.0 ± 0.5 s, 

m =	18 ± 1 g

Using the rules we discussed in class, and assuming the errors are independent, compute the following quantities (manually, i.e. without the help of a program, yet type out steps as text in Jupyter notebook) with their absolute and relative uncertainties.

1.	a + c
1.	a + b – c
1.	c * t
1.	(m * b) / t 
1.	10 * b – (a + c)
1.	exp(a*b/c^2)










## Solution

```
import numpy as np

#data
a = 5 #cm
a_err = 1 #cm
b = 1.7 #cm
b_err = 0.2 #cm
c = 12 #cm
c_err = 1 #cm
t = 3.0 #s
t_err = 0.5#s
m = 18 #g
m_err = 1 #g

# Error analysis
#1
value =  a + c
error_abs = np.sqrt((a_err)**2+(c_err)**2)
error_rel = error_abs / value*100 #%
print("1) a + c = ", a + c, "±", np.around(error_abs,1), "cm, or ±",np.around(error_rel,0), "%")

#2
value =  a + b - c
error_abs = np.sqrt((a_err)**2+(b_err)**2+(c_err)**2)
error_rel = np.abs(error_abs/value*100) #%
print("2) a + b - c = ", np.around(value,1), "±", np.around(error_abs,1), "cm, or ±",np.around(error_rel,0), "%")
      

#3
value =  c*t
error_rel = np.sqrt((c_err/c)**2+(t_err/t)**2)*100 #%
error_abs = error_rel/100*value
print("3) c * t = ", value, "±", np.around(error_abs,0), "cm*s, or ±",np.around(error_rel,0), "%")
      
#4
value =  (m * b) / t
error_rel = np.sqrt((m_err/m)**2+(b_err/b)**2+(t_err/t)**2)*100 #%
error_abs = error_rel/100*value
print("4) (m * b) / t = ", np.around(value,1), "±", np.around(error_abs,1), "cm*g/s, or ±",np.around(error_rel,0), "%")

#5 
value =  10 * b - (a + c)
error_abs = np.sqrt((a_err)**2+(c_err)**2+(10*b_err)**2)
#error_rel = error_abs / value*100 #%
print("5) 10 * b - (a + c) = ", np.around(value,1), "±", np.around(error_abs,1), "cm*g/s")
print(" NOTE: not possible to provide relative error for this case. Need to carry over absolute value in next calcuation.")

#6
value =  np.exp(a*b/c**2)
error_rel = np.sqrt((a_err/a)**2+(b_err/b)**2+(2*c_err/c)**2)*(a*b/c**2)*100 #%
error_abs = error_rel/100*value
print("6) exp(a*b/c^2) = ", np.around(value,3), "±", np.around(error_abs,3), "unitless")
```
