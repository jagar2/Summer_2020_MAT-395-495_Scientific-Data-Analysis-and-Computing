# Problem 1

1) Calculate the chi-squared value (manually and, separately, using a python function) for the following observed data from an experiment. The expected value is 15 for each measurement.

16, 18, 20, 14, 11, 12 

## Solution

```
data= [16, 18, 20, 14, 11, 12 ]
expect = np.linspace(15,15,len(data))

# Manually performing calculation
data_diff=((data-expect)/np.sqrt(expect))**2
chi_squared_manual = np.sum(data_diff)

print('Manual: ', np.around(chi_squared_manual,2))
# Using python
chi_squared,p = stats.chisquare(data,expect)
print('Function: ', np.around(chi_squared_manual,2))
```
