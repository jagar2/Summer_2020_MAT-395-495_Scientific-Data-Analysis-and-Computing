# **Homework 1**


In this Homework assignment, you will load a remote data file into a dataframe.  Note that this is a realistic data file in that it is an ASCII file but it has some header info and the delimiter is not a comma.  You'll need to figure out how to deal with that. 

The actual data is G(r) and r, that was created from x-ray scattering of an aluminum oxide thin film.  This data is a pair distribution function (PDF) representation of atoms.  Each peak in the G(r) represents a "pair distance" between two atoms.  Therefore, the first "peak" you see (after about 1.5 Angstrom) represents the closest pairs in the solid (in this case Al-O pairs).  

1.   Using the file "ALDalumina.gr" in the /Homework/HW1 folder on github, load this file into a dataframe.  It is an ASCII file, but there may be header information that you may ignore.  
2.   Plot the first column (r in Angstrom) on the x axis and the second column (G(r)) on the y axis on a linear plot.  Label axes!
3.   Calculate and plot the derivitive (dG/dr) on a second plot, as a function of r.  (there is a numpy function for this that you were not shown in class)
4.   Using python commands in your program, save a .png and a .svg image file of your plots.
